```python
def subtractFromAll(theList, theValue):
    return [x-theValue for x in theList]
```

```python
def arrangeVectorsByValues(values, vectors):
    newOrder = sorted(range(len(values)), key=lambda i: values[i-1])
    return np.matrix([vectors[x] for x in newOrder])
```

```python
def compressData(x_vals, y_vals, dimensions):
    adjustedData = np.matrix([subtractFromAll(x_vals, x_mean), subtractFromAll(y_vals, y_mean)])
    covarianceAdjData = np.cov(adjustedData)
    
    eigenValues, eigenVectors = la.eig(covarianceAdjData)
    featureVector = arrangeVectorsByValues(eigenValues, np.transpose(eigenVectors))
    result = np.dot(featureVector, adjustedData)
    
    return result, featureVector
```

```python
x_vals = [2.5, 0.5, 2.2, 1.9, 3.1, 2.3, 2.0, 1.0, 1.5, 1.1]
y_vals = [2.4, 0.7, 2.9, 2.2, 3.0, 2.7, 1.6, 1.1, 1.6, 0.9]
x_mean = np.mean(x_vals)
y_mean = np.mean(y_vals)
result, featureVector = compressData(x_vals, y_vals, 2)
```

```python
orig = np.dot(featureVector.I, result)
orig[0] = orig[0] + x_mean
orig[1] = orig[1] + y_mean
```

```python
def getListRepOfSqrImg(img):
    intensityPixels = [[col[0] for col in row] for row in img]
    result = list()
    
    for row in intensityPixels:
        result += row
        
    return result
```

```python
def convertListBackToSqrImg(l, n):
    indices = [x*250 for x in range(n)]
    return np.array([[[l[x],l[x],l[x]] for x in range(i, i+250)] for i in indices])
```

```python
def getVectorRepOfImgList(imgList, dimensionOfSqrImages):
    vectors = list()
    for x in range(dimensionOfSqrImages):
        for y in range(dimensionOfSqrImages):
            vectors += [[img[x][y][0] for img in imgList]]
    return np.array(vectors)
```
